### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

The explicit and the implicit interfaces.The recommended approach is the explicit.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

A figure represents the entire drawing area or window where one or more plots are drawn. It keeps track of all its child objects like Axes, titles, legends, and other artistic objects.

Axes is the region within the figure where the actual data is plotted. It contains the x- and y-axes and is where you can add graphs, labels, and other graphical elements.

An axis is a part of the Axes that represents one of the axes (either x or y). It is responsible for setting the scale, tick marks, and labels on the axis.

An artist is any element that can be drawn in the figure, such as lines, text, patches, or entire Axes and Figures. Almost everything you see in a plot is an artist in Matplotlib.

### When plotting in Matplotlib, what is the expected input data type?

numpy.array or numpy.ma.masked.array. These are the most common data types for input.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 100)  # Define the range for x
y = x**2  # Calculate y as x squared

plt.plot(x, y)  # Plot x and y
plt.xlabel("x label")  # Label the x-axis
plt.ylabel("y label")  # Label the y-axis
plt.title("y = x^2")  # Add a title to the plot
plt.show()  # Show the plot
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 100)
y = x**2

fig, ax = plt.subplots()  # Create a figure and an axes
ax.plot(x, y)  # Plot on the axes
ax.set_xlabel("x label")  # Add an x-label to the axes
ax.set_ylabel("y label")  # Add a y-label to the axes
ax.set_title("y = x^2")   # Add a title to the axes

plt.show()  # Show the plot
```

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Create subplot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Create scatter plot first axis
ax1.scatter(random_data_x, random_data_y, color='blue', alpha=0.5)
ax1.plot(x, y, color='red')
ax1.set_title('Scatter Plot')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')

# Create bar plot secon axis
ax2.bar(names, values, color='orange')
ax2.set_title('Bar Plot')
ax2.set_xlabel('Fruit')
ax2.set_ylabel('Quantity')

# Show plot
plt.tight_layout()
plt.show()
```
